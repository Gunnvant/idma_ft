```python
url="https://www.naukri.com/data-science-jobs"
from selenium import webdriver
browser = webdriver.Chrome()
```

```python
browser.get(url)
html=browser.page_source
```

```python
html_parsable=BeautifulSoup(html,'html.parser')
```

```python
title=[]
for i in html_parsable.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
    for j in i.find_all("div",class_="jobTupleHeader"):
        title.append(j.a.text)
```

```python
salary=[]
for i in html_parsable.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
    for j in i.find_all("div",class_="jobTupleHeader"):
        salary.append(j.find_all("li")[1].text.strip())
```

```python
exp=[]
for i in html_parsable.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
    for j in i.find_all("div",class_="jobTupleHeader"):
        exp.append(j.find_all("li")[0].text.strip())
```

```python
company=[]
for i in html_parsable.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
    for j in i.find_all("div",class_="jobTupleHeader"):
        company.append(j.find("div").find_all("a")[1].text.strip())
```

```python
loc=[]
for i in html_parsable.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
    for j in i.find_all("div",class_="jobTupleHeader"):
        loc.append(j.find_all("li")[2].text.strip())
```

```python
import pandas as pd
table=pd.DataFrame({'title':title,'salary':salary,'exp':exp,'org':company,'loc':loc})
```

```python
### Class Excercise: Scrape the information from first 4 pages
def get_data(soup):
    title=[]
    salary=[]
    exp = []
    company = []
    loc = []
    for i in soup.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
        for j in i.find_all("div",class_="jobTupleHeader"):
            title.append(j.a.text)
    for i in soup.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
        for j in i.find_all("div",class_="jobTupleHeader"):
            salary.append(j.find_all("li")[1].text.strip())
    for i in soup.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
        for j in i.find_all("div",class_="jobTupleHeader"):
            exp.append(j.find_all("li")[0].text.strip())
    for i in soup.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
        for j in i.find_all("div",class_="jobTupleHeader"):
            company.append(j.find("div").find_all("a")[1].text.strip())
    for i in soup.find_all("article",class_="jobTuple bgWhite br4 mb-8"):
        for j in i.find_all("div",class_="jobTupleHeader"):
            loc.append(j.find_all("li")[2].text.strip())
    table=pd.DataFrame()
    table['title']=title
    table['company']=company
    table['exp']=exp
    table['loc']=loc
    table['salary']=salary
    return table
```

```python
browser = webdriver.Chrome()
browser.get(url)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
get_data(soup)
```

```python
browser.close()
```

```python
from tqdm import tqdm
import time
base_url="https://www.naukri.com/data-science-jobs-{}"
master=[]
for i in tqdm(range(1,5)):
    url=base_url.format(i)
    browser = webdriver.Chrome()
    browser.get(url)
    time.sleep(1)
    html=browser.page_source
    browser.close()
    soup=BeautifulSoup(html,'html.parser')
    master.append(get_data(soup))    
```

```python
pd.concat([i for i in master])
```

```python
### Extracting links:
url="https://en.wikipedia.org/wiki/List_of_terrorist_incidents"
browser=webdriver.Chrome()
browser.get(url)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
browser.close()
```

```python
links=soup.find_all("div",class_="div-col")
```

```python
len(links)
```

```python
partial_links=[]
for link in links[0].find_all("li"):
    partial_links.append(link.a['href'])
```

```python
partial_links=pd.Series(partial_links)
```

```python
partial_links[partial_links=="/wiki/List_of_terrorist_incidents_in_2015"]
```

```python
rel_partial_links=partial_links[45:]
```

```python
base_url="https://en.wikipedia.org{}"
```

```python
def get_data(soup,i):
    tables=soup.find_all("table",class_='wikitable')
    date=[]
    typ=[]
    dead=[]
    inj=[]
    loc=[]
    details=[]
    perp=[]
    partof=[]
    for row in tables[0].find_all("tr")[1:]:
        data=row.find_all("td")
        date.append(data[0].text.strip())
        typ.append(data[1].text.strip())
        dead.append(data[2].text.strip())
        inj.append(data[3].text.strip())
        loc.append(data[4].text.strip())
        details.append(data[5].text.strip())
        perp.append(data[6].text.strip())
        try:
            partof.append(data[7].text.strip())
        except:
            partof.append("none")
    terror=pd.DataFrame()
    terror['date']=date
    terror['type']=typ
    terror['dead']=dead
    terror['inj']=inj
    terror['loc']=loc
    terror['details']=details
    terror['perp']=perp
    terror['partof']=partof
    terror['url']=i
    return terror
```

```python
master=[]
browser = webdriver.Chrome()
from tqdm import tqdm
for i in tqdm(rel_partial_links):
    url=base_url.format(i)
    browser.get(url)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    master.append(get_data(soup,i))
browser.close()
```

```python
pd.concat([i for i in master])
```

```python
big_table = pd.concat([i for i in master])
```

```python
### By, year, month,date how many terror incidents have happened
big_table['month']=big_table['url'].map(lambda x:x.split("_")[5])
big_table['year']=big_table['url'].map(lambda x:x.split("_")[-1])
big_table['country']=big_table['loc'].map(lambda x:x.split(",")[-1])
```

```python
import re
```

```python
pattern=re.compile(r'\d{1,3}')
```

```python
import numpy as np
def get_injured(row):
    if len(re.findall(pattern,row))==0:
        return np.nan
    elif len(re.findall(pattern,row))==1:
        return re.findall(pattern,row)[0]
    else:
        els=re.findall(pattern,row)
        els=list(map(lambda x:float(x),els))
        els=np.mean(els)
        return els
```
